```python
tarefas = []
escolha = 0
concluir = 0
temp = ""

while True:
    print("================gerenciador de tarefas==================")
    escolha = input("escolha uma opção\n1-adicionar tarefa\n2-marcar concluida\n3-visualizar tarefas\n4-remover tarefa\n5-encerrar")
    if escolha == 1:
        tarefas.append(input("digite a tarefa: "))
    if escolha == 2:
        concluir = int(input(f"escolha uma tarefa para concluir entre 0 e {len(tarefas) - 1}: "))
        if 0 <= concluir > len(tarefas):
            temp = tarefas[concluir]
            tarefas.pop(concluir)
            tarefas.apend(temp + " concluida")
        else:
            print("Índice inválido.")
    if escolha == 3:
        print(tarefas)
    if escolha == 4:
        concluir = int(input(f"escolha uma tarefa para excluir entre 0 e {len(tarefas) - 1}: "))
        if 0 <= concluir < len(tarefas):
            tarefas.pp(concluir)

            print("Índice inválido.")
    if escolha == 5:
        break
```
