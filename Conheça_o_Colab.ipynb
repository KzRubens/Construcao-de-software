```
def exibir_menu():
    print("\n--- GERENCIADOR DE TAREFAS ---")
    print("1. Adicionar tarefa")
    print("2. Marcar tarefa como concluída")
    print("3. Visualizar todas as tarefas")
    print("4. Remover uma tarefa")
    print("5. Encerrar o programa")
    return input("Escolha uma opção (1-5): ")

def adicionar_tarefa(lista):
    titulo = input("Digite o nome da tarefa: ")

    lista.append({"titulo": titulo, "concluida": False})
    print(f"Tarefa '{titulo}' adicionada com sucesso!")

def visualizar_tarefas(lista):
    if not lista:
        print("\nNenhuma tarefa cadastrada.")
        return

    print("\n--- SUAS TAREFAS ---")
    for indice, tarefa in enumerate(lista):
        status = "✅" if tarefa["concluida"] else "❌"
        print(f"{indice + 1}. [{status}] {tarefa['titulo']}")

def marcar_concluida(lista):
    visualizar_tarefas(lista)
    if not lista: return

    try:
        escolha = int(input("Digite o número da tarefa que concluiu: ")) - 1
        if 0 <= escolha < len(lista):
            lista[escolha]["concluida"] = True
            print(f"Tarefa '{lista[escolha]['titulo']}' marcada como concluída!")
        else:
            print("Número de tarefa inválido.")
    except ValueError:
        print("Por favor, digite um número válido.")


def remover_tarefa(lista):
    visualizar_tarefas(lista)
    if not lista: return

    try:
        escolha = int(input("Digite o número da tarefa que deseja remover: ")) - 1

        tarefa_removida = lista.pop(0)

        print(f"Tarefa '{tarefa_removida['titulo']}' removida com sucesso!")
    except ValueError:
        print("Por favor, digite um número válido.")


def executar_programa():
    tarefas = []

    while True:
        opcao = exibir_menu()

        if opcao == "1":
            adicionar_tarefa(tarefas)
        elif opcao == "2":
            marcar_concluida(tarefas)
        elif opcao == "3":
            visualizar_tarefas(tarefas)
        elif opcao == "4":
            remover_tarefa(tarefas)
        elif opcao == "5":
            print("Encerrando o programa... Até mais!")
            break
        else:
            print("Opção inválida! Tente novamente.")

# Inicia o programa
executar_programa()
```
