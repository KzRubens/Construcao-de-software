```python
tarefas = []
concluidas = []

while True:
    print("\n=== GERENCIADOR DE TAREFAS ===")
    print("1 - Adicionar tarefa")
    print("2 - Marcar tarefa como concluída")
    print("3 - Visualizar todas as tarefas")
    print("4 - Remover tarefa")
    print("5 - Encerrar programa")

    opcao = input("Escolha uma opção: ")

    if opcao == "1":
        tarefa = input("Digite a tarefa: ")

        concluidas.append(False)
        print("Tarefa adicionada!")

    elif opcao == "2":
        for i in range(len(tarefas)):
            print(i + 1, "-", tarefas[i])

        num = int(input("Número da tarefa concluída: "))

        if 1 <= num <= len(tarefas):
            concluidas[num - 1] = True
            print("Tarefa marcada como concluída!")
        else:
            print("Número inválido!")

    elif opcao == "3":
        if len(tarefas) == 0:
            print("Nenhuma tarefa cadastrada.")
        else:
            print("\nLista de tarefas:")
            for i in range(len(tarefas)):
                if concluidas[i]:
                    status = "Concluída"
                else:
                    status = "Pendente"

                print(i + 1, "-", tarefas[i], "-", status)

    elif opcao == "4":
        for i in range(len(tarefas)):
            print(i + 1, "-", tarefas[i])

        num = int(input("Número da tarefa para remover: "))

        if 1 <= num <= len(tarefas):
            tarefas.pop(num - 1)
            concluidas.pop(num - 1)
            print("Tarefa removida!")
        else:
            print("Número inválido!")

    elif opcao == "5":
        print("Programa encerrado.")
        break

    else:
        print("Opção inválida!")
```
